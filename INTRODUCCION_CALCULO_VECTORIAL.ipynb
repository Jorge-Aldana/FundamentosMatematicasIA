<h2>Ejercicio</h2>
El problema consiste en determinar:

*  
Calcular la derivada de una función de varias variables  
$f(x,y,z)=5xy-xcos(z)+x^{2}+z^{8y}$



```python
from sympy import symbols, cos, diff

a, b, c = symbols("a b c", real=True)
f = 5 * a * b - a * cos(c) + a ** 2 + c ** 8 * b

# Derivada de la función f con respecto a la variable denotada como a
print("La derevida de f respecto a x es:",diff(f, a))
# Derivada de la función respecto a la variable y
print("La derevida de f respecto a y es:", diff(f, b))
# Derivada de la función respecto a la variable z
print("La derevida de f respecto a y es:", diff(f, c))
```

 <h2>Curvas parámetricas</h2>
 Los métodos de kernel , un área importante del aprendizaje automático, pueden implicar el cálculo de funciones con valores vectoriales, que luego se pueden utilizar en el aprendizaje multitarea o en el aprendizaje por transferencia. Es por esto, que enterderlas y poder visualizarlas resultan de gran ayuda. Para esto, veamos como se grafica la curva paramétrica:

  $r(t)=(cos(t),sen(t),t)$   para $0 \leq t \leq 4\pi $  
  Aqui puede observar que el parametro toma valores en el intervalo $[0,4\pi] $ y que la componente $x(t)=cos(t)$, independientemente de los valores que tome el  parámetro esta  $[-1,1] $ como se puede apreciar en la figura.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

t = np.linspace(0, 4*np.pi, 100)
x = np.cos(t)
y = np.sin(t)
z = t

ax.plot(x, y, z)
```

```python

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
ax = plt.axes(projection='3d')

# otra forma cambiando los valores que toma el paramétro z=t
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Sobreponer puntos en la curva
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

Por su parte, para obtener una superficie en el espacio tridimensional, imagine el sistema de coordenadas  $(x,y)$
  en plano. Entonces, cada punto del dominio de la función  $f$
  tiene un único valor  $z$  asociado a él. En el siguiente código, encontrara como graficar una superficie.

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de muestra recuerde que tienen que ver con el dominio
X, Y = np.meshgrid(np.linspace(-8, 8),
                   np.linspace(-8, 8))
R = np.sqrt(X**2 + Y**2)
# la superficie z
Z = np.sin(R) / R

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

# Superficie 3D
ax.plot_surface(X, Y, Z)

# Etiquetas de los ejes
ax.set_xlabel('Etiqueta del eje X')
ax.set_ylabel('Etiqueta del eje Y')
ax.set_zlabel('Etiqueta del eje Z')

# Título
plt.title("Superficie 3D")

# plt.show()



```
